```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tt = pd.read_csv('titanic.csv')
tt
```

```python
tt.shape
```

```python
tt.head(5)
```

```python
#Checking Null Values

tt.isna().sum()
```

```python
#Removing Cabin Column as it has almost 70% null values

tt = tt.drop('Cabin', axis = 1)
```

```python
tt.isna().sum()
```

```python
#Replacing null values in AGE column with mean value

tt.Age.fillna(tt.Age.mean(), inplace=True)
```

```python
tt.isna().sum()
```

```python
# replacing the null value of embarked with mode as it has most repeated values

md = tt.Embarked.mode()[0]
tt['Embarked'].fillna(md,inplace = True)
```

```python
tt
```

```python
tt.isna().sum()
```

```python
#Dropping Columns which are unwanted
tt = tt.drop(['PassengerId','Name'], axis = 1)
```

```python
tt
```

```python
#Encoding for character columns (Sex & Embarked)
gender = pd.get_dummies(tt['Sex'], drop_first=True)

#here the dummies represents the categorical variables as binary values (0 or 1)
```

```python
Embarked = pd.get_dummies(tt.Embarked, drop_first = True)
```

```python
tt
```

```python
#Dropping Columns which are unwanted

tt.drop(['Sex','Embarked','Ticket'], axis = 1)
```

```python
tt
```

```python
#Here Concatenate (combine) two or more DataFrames

tt = pd.concat([tt,gender,Embarked], axis = 1)
```

```python
tt
```

```python
#Dropping Columns which are unwanted

tt.drop(['Sex','Ticket','Embarked'], axis = 1, inplace=True)
```

```python
tt
```
